# 4. Develop an Algorithm

Here is our pseduocode:

    Get the data
    Calculate the total consumption
    plot the results

````python
energy_data = {
    "month": [],
    "electricity": [],
    "gas": [],
    "water": []
}

for i in range(1, 4):
    month = input("\nEnter the month: ")
    energy_data["month"].append(month)

    print(f"\nEnter energy consumption data for {month}: ")

    for energy_type in ["electricity", "gas", "water"]:
        consumption = float(input(f"Enter {energy_type} consumption: "))
        energy_data[energy_type].append(consumption)

# Plot the results
total_consumption = {energy_type: sum(energy_data[energy_type]) for energy_type in ["electricity", "gas", "water"]}
plt.bar(['Electricity', 'Gas', 'Water'], [total_consumption['electricity'], total_consumption['gas'], total_consumption['water']])
plt.xlabel('Energy Type')
plt.ylabel('Total Usage')
plt.title('Comparison of Energy Types by Consumption')
plt.show()
````

# 5. Test Solution with a variety of data

The final step in our problem solving process is testing the solution.  we should first test with out hand exmaple because we have already computed the solution.

If you look above we can see the energy consumption example, and we can see the plot.  Because this matches the one form the hand example, we now replace the data with our real energy data.

````python
import matplotlib.pyplot as plt
````

````python
# define the month and three different types energy data and make a dict
energy_data = {
    "month": [],
    "electricity": [],
    "gas": [],
    "water": []
}
# set a 12 months loop and storage the data
while True:
    try:
        for i in range(1, 13):
            month = input("\nEnter the month: ")
            energy_data["month"].append(month)

            print(f"\nEnter energy consumption data for {month}: ")

            for energy_type in ["electricity", "gas", "water"]:
                consumption = float(input(f"Enter {energy_type} consumption: "))
                energy_data[energy_type].append(consumption)
        break
    except ValueError:
        print('\nSorry, that was an invalid input.\nPlease enter a numerical number')
````

````python
# make a loop to do the multiple choose
while True:
    print('\nChoose 1 for Monthly consumption trends. \nChoose 2 for Comparison of energy types by consumption. \nChoose 3 for Annual total consumption breakdown by energy type. \nEnter 0 to exit.')
    option = int(input('Enter your option: '))

    if option == 1:
        # Compare monthly consumption of three energy types
        for energy_type in ["electricity", "gas", "water"]:
            plt.plot(energy_data["month"], energy_data[energy_type], label=energy_type)

        plt.xlabel('Month')
        plt.ylabel('Consumption')
        plt.title('Monthly Energy Consumption Trends')
        plt.legend()
        plt.show()

    elif option == 2:
    # Calculate and compare total consumption between three different energy type
        total_consumption = {energy_type: sum(energy_data[energy_type]) for energy_type in ["electricity", "gas", "water"]}
        plt.bar(['Electricity', 'Gas', 'Water'], [total_consumption['electricity'], total_consumption['gas'], total_consumption['water']])
        plt.xlabel('Energy Type')
        plt.ylabel('Total Usage')
        plt.title('Comparison of Energy Types by Consumption')
        plt.show()

    elif option == 3:
        # Annual total consumption breakdown by energy type
        total_consumption = {energy_type: sum(energy_data[energy_type]) for energy_type in ["electricity", "gas", "water"]}
        plt.pie([total_consumption[energy_type] for energy_type in total_consumption], labels=total_consumption.keys(), autopct='%1.1f%%')
        plt.title('Annual Total Consumption Breakdown by Energy Type')
        plt.show()

    elif option == 0:
        print("Exiting the program.")
        break

    else:
        print("Invalid option. Please try again.")
````
